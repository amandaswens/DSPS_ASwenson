```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import numpy as np
import pylab as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
np.unique(x_train)
```

```python
x_train = x_train.astype(float) / 255.
x_test = x_test.astype(float) / 255.
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
x_train.shape
```

```python
for i in range(10):
  plt.imshow(x_train[i])
  plt.show()
```

```python
y_train[:10]
```

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
x_trainraveled = x_train.reshape(x_train.shape[0], x_train.shape[1]*x_train.shape[2]*x_train.shape[3])
```

```python
x_testraveled = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2]*x_test.shape[3])
```

```python
x_test.shape
```

```python
x_testraveled.shape
```

```python
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2]*x_test.shape[3])
```

```python
mlp = keras.Sequential([
    layers.Dense(320, activation='sigmoid'),
    layers.Dense(32, activation='sigmoid'),
    layers.Dense(32, activation='sigmoid'),
    layers.Dense(10, activation='softmax')
])
```

```python
mlp.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = mlp.fit(x_trainraveled, y_train, batch_size=120, epochs=10, validation_split=0.25, verbose=1)
```

```python
history.history
```

```python
plt.plot(history.history['loss'], label="Training")
plt.plot(history.history['val_loss'], label="Test")
plt.title('Model Loss')
plt.xlabel("Epoch")
plt.ylabel("Log_Loss")
plt.legend()
plt.yscale('log')
```

**CAPTION:** The plot above shows the loss function over multiple epochs. Blue represents the training loss and orange the test loss. It is obvious from the plot that both test and train decrease as epochs pass which is what we want. It suggests that the model is learning effectively without significant overfitting. That being said, the training data does cross the test data at around 3 epochs which could indicate a small amount of overfitting beginning to occur.

```python
mlp.evaluate(x_testraveled, y_test)
```

```python
# convolution part - i got this from tensorflow tutorial on convolution neural network

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss= 'categorical_crossentropy', metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=120, epochs=10, validation_split=0.25, verbose=1)
```

```python
history.history
```

```python
plt.plot(history.history['loss'], label="Training")
plt.plot(history.history['val_loss'], label="Test")
plt.title('Model Loss for Convolution')
plt.xlabel("Epoch")
plt.ylabel("Log_Loss")
plt.legend()
plt.yscale('log')
```

**CAPTION:** This plot shows the model loss over training epochs for a convolutional neural network. The training is in blue, and test in orange just as it was in Figure 1. The training loss decreases steadily, while the test loss levels off after the first few epochs, indicating that the model is learning effectively without major overfitting. Both losses stabilize at low values, suggesting good generalization performance.

```python
plt.plot(history.history['accuracy'], label="Training")
plt.plot(history.history['val_accuracy'], label="Test")
plt.title('Training vs. Validation Accuracy')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
```

**CAPTION:** This plot shows how the training and test accuracy improve over each epoch. Both curves plateau pretty early and stay very close to each other, which means the model is learning well and isnt really overfitting (i.e. model is a good fit).
